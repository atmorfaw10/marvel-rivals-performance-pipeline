```python
import requests
import os
from dotenv import load_dotenv
load_dotenv()

API_KEY = os.getenv("MARVEL_RIVALS_API_KEY")
HEADERS = {"x-api-key": API_KEY}
BASE = "https://marvelrivalsapi.com/api"

r = requests.get(f"{BASE}/v1/heroes", headers=HEADERS)
print("Status:", r.status_code)
print(r.text[:1000])
```

```python
import json

r = requests.get(f"{BASE}/v1/heroes", headers=HEADERS)
print(f"Status: {r.status_code}")
heroes = r.json()
print(f"Total heroes: {len(heroes)}")

# Save locally
with open("marvel_rivals_data/heroes.json", "w") as f:
    for hero in heroes:
        f.write(json.dumps(hero) + '\n')

print("Saved heroes.json")
```
